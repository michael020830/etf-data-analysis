```python
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime
from dateutil.relativedelta import relativedelta
import os
from google.colab import drive
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

pd.options.display.float_format = '{:.3f}'.format

# 탐색 및 정렬 - 이산 가중치 생성 (조합적 탐색)
def generate_discrete_weights(num_assets, increment=0.1):
    # 주어진 자산 수(num_assets)와 증분(increment)에 따라 가능한 모든 포트폴리오 가중치 조합을 생성합니다.
    # 이는 동적 프로그래밍의 '상태 공간'을 정의하는 조합적 탐색 과정입니다.
    if num_assets == 1:
        return [(1.0,)]

    weights = []

    def _generate(index, current_sum, current_weights):
        if index == num_assets - 1:
            last_weight = 1.0 - current_sum
            if 0 - 1e-9 <= last_weight <= 1.0 + 1e-9:
                weights.append(tuple(current_weights + [round(last_weight, 1)]))
            return

        start_weight = 0.0
        while current_sum + start_weight <= 1.0 + 1e-9:
            _generate(index + 1, current_sum + start_weight, current_weights + [round(start_weight, 1)])
            start_weight += increment

    _generate(0, 0.0, [])
    return [w for w in weights if abs(sum(w) - 1.0) < 1e-9]

# 통계적 성과 지표 계산 - 포트폴리오 수익률 및 분산
# 주어진 가중치, 예상 수익률, 공분산 행렬을 사용하여 포트폴리오의 수익률과 분산을 계산합니다.
def calculate_portfolio_metrics(weights, expected_returns, cov_matrix):
    weights = np.array(weights)
    portfolio_return = np.sum(weights * expected_returns)
    portfolio_variance = np.dot(weights.T, np.dot(cov_matrix, weights))
    return portfolio_return, portfolio_variance

# 리밸런싱 시 발생하는 거래 비용을 계산합니다.
def calculate_transaction_cost(current_weights, next_weights, current_capital_value, transaction_cost_rate):
    weight_diff = np.abs(next_weights - current_weights)
    total_weight_change = np.sum(weight_diff)
    cost = total_weight_change * current_capital_value * transaction_cost_rate
    return cost

# 금융 데이터를 숫자로 변환하기 위한 유틸리티 함수
def clean_financial_string(s, multiplier=1):
    if pd.isna(s):
        return np.nan
    s = str(s).replace('$', '').replace(',', '').strip()
    if 'T' in s:
        return float(s[:-1]) * 1_000_000_000_000 * multiplier
    elif 'B' in s:
        return float(s[:-1]) * 1_000_000_000 * multiplier
    elif 'M' in s:
        return float(s[:-1]) * 1_000_000 * multiplier
    elif 'K' in s:
        return float(s[:-1]) * 1_000 * multiplier
    else:
        try:
            return float(s) * multiplier
        except ValueError:
            return np.nan

print("Google Drive 마운트를 시도합니다...")
try:
    drive.mount('/content/drive')
    print("Google Drive 마운트 완료.")
except Exception as e:
    print(f"Google Drive 마운트 중 오류 발생: {e}")
    print("수동으로 마운트를 시도하거나, Colab 환경을 확인해주세요.")

BASE_DIR = '/content/drive/MyDrive/Colab_Data'
os.makedirs(BASE_DIR, exist_ok=True)
print(f"'{BASE_DIR}' 디렉토리 확인 또는 생성 완료.")

def load_data(file_path, **kwargs):
    try:
        df = pd.read_csv(file_path, **kwargs)
        return df
    except FileNotFoundError:
        return pd.DataFrame()

# 1단계: ETF 데이터 수집 & 전처리 (yfinance 사용)
etf_info_df = load_data(os.path.join(BASE_DIR, 'etf_data.csv'))
etf_price_df = load_data(os.path.join(BASE_DIR, 'etf_price.csv'), index_col=0, parse_dates=True)

if etf_info_df.empty or etf_price_df.empty:
    print("오류: 'etf_info_df' 또는 'etf_price_df' 데이터가 비어있습니다. 프로그램 종료.")
    exit()
else:
    if 'Type' in etf_info_df.columns:
        etf_info_df.rename(columns={'Type': 'Asset_Type'}, inplace=True)
        print("컬럼 'Type'을 'Asset_Type'으로 성공적으로 변경했습니다.")
    print("etf_info_df 컬럼 클리닝 및 타입 변환 중...")

    # 탐색 및 정렬 - 데이터 전처리/클리닝 (정렬에 사용될 수 있는 데이터 형식화)
    etf_info_df['AUM'] = etf_info_df['AUM'].apply(clean_financial_string)
    etf_info_df['Weighted_Average_Market_Cap'] = etf_info_df['Weighted_Average_Market_Cap'].apply(clean_financial_string)
    for col in ['Expense_Ratio', 'Dividend_Yield']:
        etf_info_df[col] = etf_info_df[col].astype(str).str.replace('%', '', regex=False).replace('None', np.nan).astype(float) / 100
    for col in ['PER', 'PBR']:
        etf_info_df[col] = pd.to_numeric(etf_info_df[col], errors='coerce')
    etf_info_df['Inception_Date'] = pd.to_datetime(etf_info_df['Inception_Date'], errors='coerce')

    print("etf_info_df 컬럼 클리닝 및 타입 변환 완료.")
    print("ETF 성과 지표 (수익률, 변동성) 계산 중...")
    all_symbols = etf_info_df['Symbol'].tolist()
    end_date_full = datetime.now()
    start_date_full = end_date_full - relativedelta(years=5)

    print(f"전체 ETF 가격 데이터 로드 기간: {start_date_full.strftime('%Y-%m-%d')} ~ {end_date_full.strftime('%Y-%m-%d')}")
    try:
        etf_price_df = yf.download(all_symbols, start=start_date_full, end=end_date_full)['Close']
        if isinstance(etf_price_df, pd.Series):
            etf_price_df = etf_price_df.to_frame()
        etf_price_df = etf_price_df.dropna(axis=1, how='all').dropna(axis=0)
        print(f"총 {len(etf_price_df.columns)}개의 ETF 가격 데이터 로드 완료.")

    except Exception as e:
        print(f"오류: yfinance로 ETF 가격 데이터를 로드하는 데 실패했습니다: {e}")
        print("프로그램을 종료합니다.")
        exit()

    if etf_price_df.empty:
        print("오류: 가격 데이터 로드 후에도 'etf_price_df'가 비어있습니다. 프로그램 종료.")
        exit()
    #통계적 성과 지표 계산- 일별 수익률, 연환산 수익률, 변동성
    daily_returns = etf_price_df.pct_change()
    annual_returns = (1 + daily_returns).apply(
        lambda x: x.dropna().prod()**(252/len(x.dropna())) - 1 if len(x.dropna()) > 0 else np.nan
    ).rename('Annual_Return')
    volatility = daily_returns.std() * np.sqrt(252)
    volatility.name = 'Volatility'

    print("ETF 성과 지표 계산 완료.")
    performance_metrics_df = pd.DataFrame({'Annual_Return': annual_returns, 'Volatility': volatility})
    performance_metrics_df.index.name = 'Symbol'
    # 탐색 및 정렬 - 데이터 결합 및 결측치 제거 (이후 필터링 및 정렬을 위한 준비)
    etf_combined_df = etf_info_df.set_index('Symbol').join(performance_metrics_df)
    etf_combined_df = etf_combined_df.dropna(subset=['Annual_Return', 'Volatility'])
    if etf_combined_df.empty:
        print("오류: ETF 정보와 성과 지표 통합 후 데이터가 비어있습니다. 프로그램 종료.")
        exit()


    etf_combined_data_path = os.path.join(BASE_DIR, 'etf_combined_data.csv')
    etf_combined_df.to_csv(etf_combined_data_path)
    print(f"\n'{etf_combined_data_path}' 파일로 통합 데이터가 저장되었습니다.")

    print("\n--- ETF 추천을 위한 사용자 입력 ---")
    print("아래 질문에 답변해 주세요. (괄호 안의 예시/가이드를 참고하세요.)\n")

    # 사용자 입력은 이후 ETF를 '필터링'하는 데 사용됩니다.
    investment_goal = input("""
    1. 주요 투자 목표를 선택해 주세요 (성장, 배당/수익, 안정, 혼합):
    -성장 (높은 수익률 추구, 변동성 감수)
    -배당/수익 (정기적인 분배금/배당금 선호)
    -안정 (낮은 변동성 선호, 자산 보존)
    -혼합 (성장과 안정의 균형)
    """).strip()

    preferred_themes_input = input("""
    2. 선호하는 ETF 테마가 있다면 선택해 주세요 (복수 선택 시 쉼표로 구분, 없으면 '상관 없음'):
    테마(Theme)란? 특정 산업, 투자 전략, 또는 거시 경제 트렌드를 따르는 ETF 그룹입니다. 특정 분야에 집중 투자하여 높은 수익을 추구할 수 있지만, 해당 분야의 리스크에 더 크게 노출될 수 있습니다.
    -Growth & Tech (성장 및 기술): 높은 성장 잠재력을 가진 기술 기업 등에 투자
    -Commodities & Alternatives (원자재 및 대체 투자): 금, 석유, 농산물 등 원자재 또는 헤지펀드 같은 대체 자산에 투자
    -International Equity (Developed) (선진국 국제 주식): 미국을 제외한 선진국 시장의 주식에 투자
    -US Broad Equity (미국 광범위 주식): 미국 전체 또는 넓은 범위의 주식 시장에 분산 투자
    -Sector Equity (Healthcare) (헬스케어 섹터 주식): 제약, 바이오, 의료 기기 등 헬스케어 산업에 집중 투자
    -Fixed Income (Broad) (광범위 채권): 다양한 종류의 채권에 분산 투자하여 안정적인 수익 추구
    -Fixed Income (Treasury) (국채 채권): 주로 미국 국채 등 정부 발행 채권에 투자하여 높은 안정성 추구
    -Global Broad Equity (글로벌 광범위 주식): 전 세계 다양한 국가의 주식 시장에 분산 투자
    -US Small Cap Equity (미국 소형주 주식): 미국 중소형 기업 주식에 투자, 성장 잠재력 높음
    -Real Estate (REITs) (부동산 투자 신탁): 부동산 관련 자산(주로 리츠)에 투자하여 임대 수익 및 자산 가치 상승 추구
    ★초보자 추천 : US Broad Equity,Global Broad Equity,Fixed Income (Broad)★
    """).strip()
    preferred_themes = [t.strip() for t in preferred_themes_input.split(',')] if preferred_themes_input.lower() != '상관 없음' else []

    preferred_asset_types_input = input("""
    3. 선호하는 자산 유형이 있다면 선택해 주세요 (복수 선택 시 쉼표로 구분, 없으면 '상관 없음'):
    자산 유형이란? 투자의 가장 기본적인 분류 기준으로, 각 유형마다 고유한 위험-수익 특성을 가집니다.
    주요 자산 유형과 설명:
    -주식 (Equity): 기업의 소유권을 나타내며, 높은 성장 가능성과 함께 높은 가격 변동성(위험)을 가집니다.
    -채권 (Fixed Income): 정부나 기업에 돈을 빌려주고 정기적으로 이자를 받는 형태의 투자입니다. 주식보다 상대적으로 안정적이며, 예측 가능한 수익을 기대할 수 있습니다.
    -원자재 (Commodities): 석유, 금, 농산물 등 실물 상품에 투자합니다. 인플레이션 헤지 수단으로 활용되기도 하며, 공급과 수요에 따라 가격 변동성이 높을 수 있습니다.
    -부동산 (Real Estate): 부동산 자체 또는 부동산 관련 기업(REITs 등)에 투자합니다. 임대 수익과 자산 가치 상승을 목표로 하며, 실물 자산의 특성을 가집니다.
    -혼합 (Mixed): 여러 자산 유형(예: 주식과 채권)에 분산 투자하여 위험을 줄이고 균형적인 수익을 추구하는 방식입니다.
    ★초보자 추천 : 혼합, 주식, 채권★
    """).strip()
    preferred_asset_types = [a.strip() for a in preferred_asset_types_input.split(',')] if preferred_asset_types_input.lower() != '상관 없음' else []

    try:
        min_aum_input = float(input("""
    4. 최소 운용자산(AUM) 규모를 입력해 주세요 (단위: 억 달러, 숫자로만 입력. 예: 10 -> 10억 달러 이상. 없으면 0 입력):
    AUM은 ETF가 운용하는 자산의 총 규모입니다.
    AUM이 클수록 ETF의 유동성이 높아 매매가 용이하며, 운용의 안정성이 높고 폐지될 위험이 낮아집니다.
    일반적으로 10억 달러(1,000M) 이상의 AUM을 가진 ETF는 안정적이라고 봅니다.
    ★초보자 추천 : 5~10(5억 달러 ~ 10억 달러 이상)★
        """).strip())
    except ValueError:
        print("유효하지 않은 AUM 값입니다. 0으로 설정합니다.")
        min_aum_input = 0.0

    try:
        max_expense_ratio_input = float(input("""
    5. 허용 가능한 운용보수(Expense Ratio)의 최대치를 입력해 주세요 (퍼센트 단위, 숫자로만 입력. 예: 0.5 -> 0.5% 이하. 없으면 1.0 입력):
    운용보수는 ETF를 운용하는 데 드는 비용으로, 투자 수익에서 매년 차감됩니다. 이 보수가 낮을수록 투자자의 실질 수익률이 높아집니다.
    특히 시장 지수를 추종하는 수동형(Passive) ETF의 경우, 0.2% 미만은 매우 낮은 수준, 0.2% ~ 0.5%는 일반적인 수준, 0.5% 이상은 다소 높은 수준으로 볼 수 있습니다. 액티브 ETF는 더 높을 수 있습니다.
    ★초보자 추천 : 0.3 이하 (최대한 낮은 운용보수)★
        """).strip())
    except ValueError:
        print("유효하지 않은 운용보수 값입니다. 1.0으로 설정합니다.")
        max_expense_ratio_input = 1.0

    print("\n--- ETF 필터링 진행 중 ---")
    filtered_etfs = etf_combined_df.copy()
    # 탐색 및 정렬 - 데이터 선별 (필터링)
    # 사용자의 선호도(테마, 자산 유형, AUM, 운용보수)에 따라 DataFrame을 필터링합니다.
    if preferred_themes:
        filtered_etfs = filtered_etfs[filtered_etfs['Theme'].isin(preferred_themes)]
        print(f"테마 필터링 완료: {len(filtered_etfs)}개 ETF 남음.")
    else:
        print("테마가 지정되지 않아 테마 필터링을 건너뜀.")

    if preferred_asset_types:
        if 'Asset_Type' in filtered_etfs.columns:
            filtered_etfs = filtered_etfs[filtered_etfs['Asset_Type'].isin(preferred_asset_types)]
            print(f"자산 유형 필터링 완료: {len(filtered_etfs)}개 ETF 남음.")
        else:
            print("오류: 'Asset_Type' 컬럼을 찾을 수 없어 자산 유형 필터링을 건너뜠습니다. 'etf_data.csv'에 'Type' 컬럼이 있는지 확인해주세요.")
    else:
        print("자산 유형이 지정되지 않아 자산 유형 필터링을 건너뜜.")

    if min_aum_input > 0:
        filtered_etfs = filtered_etfs[filtered_etfs['AUM'] >= min_aum_input * 1_000_000_000]
        print(f"AUM 필터링 완료 (최소 {min_aum_input}억 달러): {len(filtered_etfs)}개 ETF 남음.")
    else:
        print("AUM 최소 기준이 없어 AUM 필터링을 건너뜀.")

    if max_expense_ratio_input < 1.0:
        filtered_etfs = filtered_etfs[filtered_etfs['Expense_Ratio'] <= max_expense_ratio_input]
        print(f"운용보수 필터링 완료 (최대 {max_expense_ratio_input:.3f}%): {len(filtered_etfs)}개 ETF 남음.")
    else:
        print("운용보수 최대 기준이 없어 운용보수 필터링을 건너뜜.")

    if filtered_etfs.empty:
        print("\n선택하신 기준으로 필터링된 ETF가 없습니다. 기준을 완화하여 다시 시도해 주세요.")
        exit()
    else:
        print(f"\n--- 필터링 결과: {len(filtered_etfs)}개의 ETF가 조건에 부합합니다. ---")
        print("--- ETF 순위화 진행 중 ---")

        sort_criteria = {
            '성장': (['Annual_Return', 'Volatility'], [False, True]),
            '배당/수익': (['Dividend_Yield', 'Volatility'], [False, True]),
            '안정': (['Volatility', 'Annual_Return'], [True, False]),
            '혼합': (['Annual_Return', 'Volatility'], [False, True])
        }
        # 탐색 및 정렬 - 우선순위 지정 (정렬)
        # 투자 목표에 따라 ETF를 특정 성과 지표(Annual_Return, Volatility 등) 기준으로 정렬합니다.
        sort_by, ascending = sort_criteria.get(investment_goal, (['Annual_Return'], [False]))
        sorted_etfs = filtered_etfs.sort_values(by=sort_by, ascending=ascending, na_position='last')
        print(f"투자 목표 '{investment_goal}'에 따라 {sort_by} 기준으로 순위화.")

        display_columns = ['Name', 'Theme', 'Asset_Type', 'Expense_Ratio', 'AUM', 'Annual_Return', 'Volatility', 'Dividend_Yield', 'PER', 'PBR']
        display_columns = [col for col in display_columns if col in sorted_etfs.columns]
        print("\n--- 메인 추천 ETF ---")
        main_etf = sorted_etfs.head(1)
        print(main_etf)

        if not main_etf.empty:
            main_etf_symbol = main_etf.index[0]
            print("\n--- 시너지 ETF 탐색 중 ---")
            all_symbols_for_correlation_full = etf_combined_df.index.tolist()
            if main_etf_symbol in all_symbols_for_correlation_full:
                all_symbols_for_correlation_full.remove(main_etf_symbol)

            symbols_to_download = [main_etf_symbol] + all_symbols_for_correlation_full
            symbols_to_download = list(set(symbols_to_download))

            end_date_corr = datetime.now()
            start_date_corr = end_date_corr - relativedelta(years=1)

            print(f"상관계수 계산을 위한 데이터 로드 기간: {start_date_corr.strftime('%Y-%m-%d')} ~ {end_date_corr.strftime('%Y-%m-%d')}")

            synergy_etfs = pd.DataFrame()
            try:
                corr_price_data = yf.download(symbols_to_download, start=start_date_corr, end=end_date_corr, progress=False)['Close']
                if isinstance(corr_price_data, pd.Series):
                    corr_price_data = corr_price_data.to_frame()
                corr_price_data = corr_price_data.dropna(axis=1, how='all').dropna(axis=0)

                if not corr_price_data.empty and main_etf_symbol in corr_price_data.columns:
                  # 통계적 성과 지표 계산 - 상관 행렬
                    daily_returns_corr = corr_price_data.pct_change().dropna()
                    correlation_matrix = daily_returns_corr.corr()

                    if main_etf_symbol in correlation_matrix.columns:
                        main_etf_corr_series = correlation_matrix[main_etf_symbol]
                        potential_synergy_candidates = etf_combined_df.drop(index=main_etf_symbol, errors='ignore').copy()

                        if not potential_synergy_candidates.empty:
                          # 탐색 및 정렬 - 데이터 선별 (상관관계 필터링)
                            potential_synergy_candidates['Correlation_with_Main'] = potential_synergy_candidates.index.map(main_etf_corr_series)
                            potential_synergy_candidates['Correlation_with_Main'].fillna(1.0, inplace=True)

                            correlation_threshold = 0.5

                            filtered_synergy_by_corr = potential_synergy_candidates[
                                (potential_synergy_candidates['Correlation_with_Main'] <= correlation_threshold) |
                                (potential_synergy_candidates['Correlation_with_Main'] < 0)
                            ]

                            if filtered_synergy_by_corr.empty:
                                print(f"상관관계 임계치 ({correlation_threshold:.2f}) 이하인 시너지 ETF를 찾을 수 없습니다. 모든 후보군에서 시너지 점수 기준으로 추천을 시도합니다.")

                                potential_synergy_candidates_for_score = etf_combined_df.drop(index=main_etf_symbol, errors='ignore').copy()
                                potential_synergy_candidates_for_score['Correlation_with_Main'] = potential_synergy_candidates_for_score.index.map(main_etf_corr_series)
                                potential_synergy_candidates_for_score['Correlation_with_Main'].fillna(1.0, inplace=True)
                                potential_synergy_candidates = potential_synergy_candidates_for_score
                                print(f"경고: 상관관계 임계치 {correlation_threshold:.2f}를 만족하는 ETF가 없어, 모든 후보군에서 시너지 점수 기준으로 추천합니다. 분산 효과가 기대보다 낮을 수 있습니다.")
                            else:
                                potential_synergy_candidates = filtered_synergy_by_corr

                            main_asset_type = main_etf.loc[main_etf_symbol, 'Asset_Type'] if 'Asset_Type' in main_etf.columns else None
                            main_theme = main_etf.loc[main_etf_symbol, 'Theme'] if 'Theme' in main_etf.columns else None
                            main_region = main_etf.loc[main_etf_symbol, 'Region'] if 'Region' in main_etf.columns else None
                            # 탐색 및 정렬 - 시너지 점수 계산 (사용자 정의 정렬 기준)
                            def calculate_synergy_score(row):
                                # 메인 ETF와의 상관관계, 자산 유형/테마/지역의 차이점을 기반으로 시너지 점수를 계산합니다.
                                # 점수가 높을수록 분산 효과가 크다고 판단합니다.
                                score = 0
                                score += (1 - row['Correlation_with_Main']) * 10
                                if main_asset_type and row['Asset_Type'] != main_asset_type:
                                    score += 5
                                if main_theme and row['Theme'] != main_theme:
                                    score += 3
                                if main_region and 'Region' in row.index and row['Region'] != main_region:
                                    score += 2
                                return score

                            if not potential_synergy_candidates.empty:
                                potential_synergy_candidates['Synergy_Score'] = potential_synergy_candidates.apply(calculate_synergy_score, axis=1)
                                synergy_etfs = potential_synergy_candidates.sort_values(by='Synergy_Score', ascending=False).head(2)
                            # 탐색 및 정렬 - 시너지 점수 기준 정렬
                            else:
                                print("시너지 ETF 점수를 계산할 후보군이 없습니다.")
                        else:
                            print("시너지 ETF로 고려할 다른 ETF가 없습니다.")
                    else:
                        print(f"경고: {main_etf_symbol}에 대한 상관계수 데이터를 찾을 수 없습니다. 시너지 ETF 추천을 건너뜜.")
                else:
                    print("상관계수 계산을 위한 충분한 가격 데이터를 가져오지 못했습니다. 시너지 ETF 추천을 건너뜜.")
            except Exception as e:
                print(f"yfinance 데이터 로드 또는 시너지 ETF 탐색 중 오류 발생: {e}")

            print("\n--- 최종 추천 포트폴리오 (메인 1개 + 시너지 2개) ---")
            final_recommended_etfs = pd.concat([main_etf, synergy_etfs])

            if not final_recommended_etfs.empty:
                print("고객님의 포트폴리오 분산 및 안정성 강화를 위한 추천 조합입니다.")
                print(final_recommended_etfs[display_columns])

                print("\n--- 추천 ETF 상세 설명 ---")
                for symbol, row in final_recommended_etfs.iterrows():
                    print(f"\nETF: {row['Name']} ({symbol})")
                    print(f"  테마: {row['Theme']}, 자산 유형: {row['Asset_Type']}")
                    print(f"  연간 수익률: {row['Annual_Return']:.2%}, 연간 변동성: {row['Volatility']:.2%}")

                    if symbol == main_etf_symbol:
                        print("  **이 ETF는 고객님의 투자 목표에 가장 부합하는 핵심 ETF입니다.**")
                        print(f"  주요 투자 대상은 {row['Asset_Type']} 섹터의 {row['Theme']} 관련 자산입니다. 이 ETF는 지난 1년간 {row['Annual_Return']:.2%}의 연환산 수익률과 {row['Volatility']:.2%}의 변동성을 보였습니다.")
                    elif 'Correlation_with_Main' in row.index:
                        corr_val = row['Correlation_with_Main']

                        corr_desc = ""
                        corr_effect = ""
                        if corr_val <= 0.2:
                            corr_desc = "매우 낮은 상관관계"
                            corr_effect = f"시장 변동 시 {main_etf_symbol}의 움직임과 독립적인 경향이 강하여 포트폴리오의 안정성을 크게 높여줄 수 있습니다."
                        elif corr_val <= 0.4:
                            corr_desc = "낮은 상관관계"
                            corr_effect = f"{main_etf_symbol}과 다른 움직임을 보여 포트폴리오 전반의 위험을 분산하는 데 기여할 수 있습니다."
                        elif corr_val <= 0.6:
                            corr_desc = "보통 수준의 상관관계"
                            corr_effect = f"{main_etf_symbol}과 완전히 독립적이지는 않지만, 그래도 포트폴리오 다양성을 확보하는 데 도움이 될 수 있습니다."
                        else:
                            corr_desc = "비교적 높은 상관관계"
                            corr_effect = f"{main_etf_symbol}과 유사한 움직임을 보일 수 있어 분산 효과는 제한적일 수 있습니다. 하지만 다른 자산 유형/테마/지역 측면에서의 다양성을 제공할 수 있습니다."

                        print(f"  **이 ETF는 {main_etf_symbol}과의 {corr_desc} ({corr_val:.2f})를 통해 포트폴리오의 분산 효과를 높여줍니다.**")
                        has_different_type = False
                        if main_asset_type and row['Asset_Type'] != main_asset_type:
                            print(f"  특히, {main_etf_symbol}이 주로 **{main_asset_type}**에 투자하는 반면, 이 ETF는 **{row['Asset_Type']}**에 투자하여 **다른 자산 유형으로 위험을 분산**시킵니다.")
                            has_different_type = True
                        if not has_different_type and main_theme and row['Theme'] != main_theme:
                            print(f"  특히, {main_etf_symbol}이 주로 **{main_theme}** 테마에 투자하는 반면, 이 ETF는 **{row['Theme']}** 테마에 투자하여 **다른 테마로 위험을 분산**시킵니다.")

                        if main_region and 'Region' in row.index and row['Region'] != main_region:
                            print(f"  또한, {main_etf_symbol}이 {main_region}에 집중된 반면, 이 ETF는 {row['Region']}에 투자하여 **지역적인 분산 효과**도 얻을 수 있습니다.")
                        if not has_different_type and \
                            (not main_theme or row['Theme'] == main_theme) and \
                            (not main_region or 'Region' not in row.index or row['Region'] == main_region):
                            print(f"  이 ETF는 {main_etf_symbol}과 동일한 자산 유형/테마를 가지지만, 세부적인 투자 대상의 차이를 통해 포트폴리오의 다양성을 확보합니다.")

                        print(f"  {corr_effect}")
                    else:
                        print(f"  시너지 ETF이지만 {main_etf_symbol}과의 상관관계 정보를 가져올 수 없어 일반적인 분산 효과를 설명합니다.")
            else:
                print("최종 추천할 ETF가 없습니다. 필터링 기준을 완화하거나 다시 시도해 주세요.")
                exit()
        else:
            print("순위화된 ETF가 없어 메인 추천 ETF를 찾을 수 없습니다. 프로그램 종료.")
            exit()
    top_3_etfs = final_recommended_etfs.copy()
    recommended_etfs_path = os.path.join(BASE_DIR, 'recommended_etfs.csv')
    top_3_etfs.to_csv(recommended_etfs_path)
    print(f"\n'{recommended_etfs_path}' 파일로 최종 추천 ETF 목록이 저장되었습니다.")
print("\n" + "="*50)
print("=== 3단계: 동적 프로그래밍 기반 포트폴리오 최적화 시뮬레이션 ===")
print("="*50 + "\n")
if not top_3_etfs.empty and len(top_3_etfs) >= 1:
    print("\n--- 동적 프로그래밍을 위한 추가 정보 입력 ---")
    try:
        investment_horizon = int(input("투자 기간을 입력해 주세요 (단위: 개월, 정수로 입력. 예: 12): ").strip())
        initial_capital = float(input("초기 투자 금액을 입력해 주세요 (숫자로만 입력. 예: 10000): ").strip())
        while True:
            try:
                risk_aversion = float(input("위험 회피 계수를 입력해 주세요 (권장: 0.1 ~ 5.0 사이의 숫자, 높을수록 안정 지향): ").strip())
                if 0.0 < risk_aversion <= 10.0:
                    break
                else:
                    print("유효한 위험 회피 계수 범위를 벗어났습니다. 0.1 ~ 10.0 사이의 값을 입력해주세요.")
            except ValueError:
                print("유효하지 않은 입력입니다. 숫자로 입력해주세요.")
    except ValueError:
        print("유효하지 않은 입력입니다. 투자 기간은 12개월, 초기 투자 금액은 10000으로 기본 설정합니다.")
        investment_horizon = 12
        initial_capital = 10000.0

    risk_free_rate = 0.04
    transaction_cost_rate = 0.001

    etf_symbols = top_3_etfs.index.tolist()
    print(f"\n선택된 상위 ETF 심볼: {etf_symbols}")

    print("\n--- yfinance에서 ETF 과거 데이터 가져오는 중... ---")
    end_date = datetime.now()
    start_date = end_date - relativedelta(years=10) # 과거 10년간의 데이터

    try:
        # yfinance를 사용하여 과거 가격 데이터 다운로드 (1단계: ETF 데이터 수집)
        historical_prices = yf.download(etf_symbols, start=start_date, end=end_date)['Close']
        if isinstance(historical_prices, pd.Series):
            historical_prices = historical_prices.to_frame()

        if historical_prices.empty or historical_prices.shape[1] != len(etf_symbols):
            print("오류: yfinance에서 과거 가격 데이터를 가져오는 데 실패했거나, 모든 심볼에 대한 데이터를 가져오지 못했습니다. 심볼을 확인하거나 인터넷 연결 상태를 확인해주세요.")
            historical_data_available = False
        else:
            # 3. 통계적 성과 지표 계산 - 월별 로그 수익률, 연환산 통계량, 공분산/상관 행렬
            monthly_prices = historical_prices.resample('M').last().dropna() # 월별 최종 가격으로 리샘플링
            log_returns_monthly = np.log(monthly_prices / monthly_prices.shift(1)).dropna() # 월별 로그 수익률 계산

            trading_months_per_year = 12
            annualized_mean_returns = log_returns_monthly.mean() * trading_months_per_year # 연환산 평균 수익률
            annualized_volatilities = log_returns_monthly.std() * np.sqrt(trading_months_per_year) # 연환산 변동성
            covariance_matrix = log_returns_monthly.cov() * trading_months_per_year # 연환산 공분산 행렬
            correlation_matrix = log_returns_monthly.corr() # 상관 행렬

            print("\n--- 과거 데이터 분석 결과 (연간 기준) ---")
            print("ETF별 예상 수익률 (로그 수익률 기반):")
            print(annualized_mean_returns.apply(lambda x: f'{x:.3f}'))
            print("\nETF별 예상 변동성 (로그 수익률 기반):")
            print(annualized_volatilities.apply(lambda x: f'{x:.3f}'))
            print("\nETF들 간의 상관 행렬:")
            print(correlation_matrix.map(lambda x: f'{x:.3f}'))
            historical_data_available = True

    except Exception as e:
        print(f"yfinance 데이터 로드 중 오류 발생: {e}")
        historical_data_available = False

    if historical_data_available:
        print("\n--- 동적 프로그래밍 모델 파라미터 ---")
        print(f"투자 기간: {investment_horizon} 개월")
        print(f"초기 투자 금액: {initial_capital:.2f}")
        print(f"무위험 수익률 (연): {risk_free_rate:.3f}")
        print(f"거래 비용 (거래 금액 대비): {transaction_cost_rate:.3f}")
        print(f"위험 회피 계수: {risk_aversion:.1f}")
        print("선택된 ETF 예상 수익률 (연):")
        print(annualized_mean_returns.apply(lambda x: f'{x:.3f}'))
        print("선택된 ETF 예상 변동성 (연):")
        print(annualized_volatilities.apply(lambda x: f'{x:.3f}'))
        print("선택된 ETF 상관 행렬:")
        print(correlation_matrix.map(lambda x: f'{x:.3f}'))
        print("\n이제 이 파라미터들을 가지고 동적 프로그래밍 알고리즘을 설계할 준비가 완료되었습니다.")

        num_assets = len(etf_symbols)
        # 4. 탐색 및 정렬 - 이산 가중치 생성 (generate_discrete_weights 함수 사용)
        # 동적 프로그래밍의 상태 공간을 정의하는 부분입니다.
        possible_weights = generate_discrete_weights(num_assets, increment=0.1)

        # 1. 동적 프로그래밍 - DP 테이블 초기화
        # dp_value[t][state_idx]: t 시점에서 state_idx 가중치 상태일 때 얻을 수 있는 최대 예상 효용
        # policy[t][state_idx]: t 시점에서 state_idx 가중치 상태일 때 다음 시점으로 전이할 최적의 가중치 상태 인덱스
        dp_value = np.full((investment_horizon + 1, len(possible_weights)), -np.inf)
        policy = np.full((investment_horizon + 1, len(possible_weights)), -1, dtype=int)

        # 1. 동적 프로그래밍 - 최종 시점 (Terminal State)의 가치 초기화
        # 투자 기간 마지막 월의 가치를 계산합니다. (t=investment_horizon)
        for i, w_tuple in enumerate(possible_weights):
            # 3. 통계적 성과 지표 계산 (calculate_portfolio_metrics 함수 사용)
            ret, var = calculate_portfolio_metrics(w_tuple, annualized_mean_returns, covariance_matrix)
            # 최종 시점의 효용은 이후의 결정이 없으므로, 현재 포트폴리오의 효용으로만 결정됩니다.
            # 월별 효용으로 변환 (연환산/12)
            dp_value[investment_horizon][i] = (ret / 12) - risk_aversion * (var / 12)

        print("\n[3단계] 동적 프로그래밍 모델 계산 시작 (역방향 진행)...")
        # 1. 동적 프로그래밍 - 역방향 계산 (Backward Induction)
        # 투자 기간의 마지막 월부터 시작하여 첫 월(0)까지 시간을 거꾸로 거슬러 올라가며 계산합니다.
        for t in range(investment_horizon - 1, -1, -1): # t는 현재 시점
            print(f"   기간 {t+1} (월) 계산 중...")
            for current_weights_idx, current_w_tuple in enumerate(possible_weights): # 현재 시점의 모든 가능한 가중치 상태
                max_expected_utility = -np.inf
                best_next_weights_idx = -1
                current_w_array = np.array(current_w_tuple)

                # 모든 가능한 다음 가중치 상태(next_w_tuple)를 탐색하여 최적의 결정 찾기
                for next_weights_idx, next_w_tuple in enumerate(possible_weights):
                    next_w_array = np.array(next_w_tuple)

                    # 3. 통계적 성과 지표 계산 (calculate_transaction_cost 함수 사용)
                    # 거래 비용 계산 (현재 자본을 1.0으로 가정하여 비율로 계산)
                    cost = calculate_transaction_cost(current_w_array, next_w_array, 1.0, transaction_cost_rate)

                    # 3. 통계적 성과 지표 계산 (calculate_portfolio_metrics 함수 사용)
                    # 이 기간 동안의 포트폴리오 수익률 및 분산 계산
                    ret_this_period, var_this_period = calculate_portfolio_metrics(next_w_array, annualized_mean_returns, covariance_matrix)

                    # 현재 기간의 효용 (수익률 - 위험회피계수 * 분산 - 거래 비용)
                    utility_this_period = (ret_this_period / 12) - risk_aversion * (var_this_period / 12) - cost

                    # 다음 시점의 최대 예상 효용 (dp_value 테이블에서 조회)
                    expected_future_utility = dp_value[t+1][next_weights_idx]

                    # 총 예상 효용: 즉각적인 효용 + 다음 시점의 최대 예상 효용 (DP의 핵심 점화식)
                    total_expected_utility = utility_this_period + expected_future_utility

                    if total_expected_utility > max_expected_utility:
                        max_expected_utility = total_expected_utility
                        best_next_weights_idx = next_weights_idx

                dp_value[t][current_weights_idx] = max_expected_utility
                policy[t][current_weights_idx] = best_next_weights_idx

        print("\n[3단계] 동적 프로그래밍 최적 정책 계산 완료.")

        # 1. 동적 프로그래밍 - 최적 정책 시뮬레이션 (Forward Pass)
        # 계산된 policy 테이블을 사용하여 실제 시간 순서대로 포트폴리오 자본 변화를 시뮬레이션합니다.
        current_capital = initial_capital
        initial_weights_array = np.full(num_assets, 1.0 / num_assets)
        initial_weights_tuple_rounded = tuple(np.round(initial_weights_array, 1))

        # 4. 탐색 및 정렬 - 근접 값 탐색 (초기 가중치 인덱스 찾기)
        # 초기 가중치 배열과 가장 유사한 possible_weights 내의 인덱스를 찾습니다.
        try:
            current_weights_idx = possible_weights.index(initial_weights_tuple_rounded)
        except ValueError:
            min_diff = float('inf')
            closest_idx = -1
            for idx, pw in enumerate(possible_weights):
                diff = np.sum(np.abs(np.array(pw) - initial_weights_array))
                if diff < min_diff:
                    min_diff = diff
                    closest_idx = idx
            current_weights_idx = closest_idx

        current_portfolio_weights = np.array(possible_weights[current_weights_idx])

        capital_path_dp_sim = [initial_capital]
        portfolio_returns_simulated = []
        portfolio_variances_simulated = []

        print("\n[3단계] 최적 포트폴리오 시뮬레이션 (월별 리밸런싱):")
        for t in range(investment_horizon):
            optimal_next_w_idx = policy[t][current_weights_idx] # policy 테이블에서 다음 최적 가중치 인덱스 조회

            if optimal_next_w_idx == -1:
                print(f"경고: 기간 {t+1}에 최적 정책을 찾을 수 없습니다. 직전 가중치를 유지합니다.")
                optimal_next_w = current_portfolio_weights
            else:
                optimal_next_w = np.array(possible_weights[optimal_next_w_idx]) # 최적 가중치 배열

            # 3. 통계적 성과 지표 계산 (calculate_transaction_cost 함수 사용)
            cost_amount = calculate_transaction_cost(current_portfolio_weights, optimal_next_w, current_capital, transaction_cost_rate)
            current_capital -= cost_amount # 거래 비용만큼 자본 감소

            # 3. 통계적 성과 지표 계산 (calculate_portfolio_metrics 함수 사용)
            # 최적 가중치에 따른 월별 수익률과 분산 계산
            monthly_return_val, monthly_variance_val = calculate_portfolio_metrics(optimal_next_w, annualized_mean_returns, covariance_matrix)
            current_capital *= (1 + (monthly_return_val / 12)) # 월별 수익률 적용하여 자본 증가

            capital_path_dp_sim.append(current_capital)
            portfolio_returns_simulated.append(monthly_return_val / 12)
            portfolio_variances_simulated.append(monthly_variance_val / 12)

            current_weights_idx = optimal_next_w_idx if optimal_next_w_idx != -1 else current_weights_idx
            current_portfolio_weights = optimal_next_w

            print(f"   기간 {t+1} (월): 시작 자본: {capital_path_dp_sim[-2]:.2f}, 적용 가중치: {optimal_next_w}, "
                            f"거래 비용: {cost_amount:.2f}, 종료 예상 자본: {current_capital:.2f}")

        final_capital_dp_sim = capital_path_dp_sim[-1]
        total_return_dp_sim = (final_capital_dp_sim - initial_capital) / initial_capital

        # 3. 통계적 성과 지표 계산 - 최종 성과 지표 (수익률, 변동성, 샤프 비율)
        cumulative_monthly_return_dp_sim = np.prod([1 + r for r in portfolio_returns_simulated]) - 1
        if investment_horizon > 0 and len(portfolio_returns_simulated) > 0:
            avg_annual_return_dp_sim = (1 + cumulative_monthly_return_dp_sim)**(12/investment_horizon) - 1 # 연환산 수익률
        else:
            avg_annual_return_dp_sim = 0.0

        if len(portfolio_variances_simulated) > 0:
            avg_monthly_variance_dp_sim = np.mean(portfolio_variances_simulated)
            avg_annual_volatility_dp_sim = np.sqrt(avg_monthly_variance_dp_sim * 12) if avg_monthly_variance_dp_sim >= 0 else 0 # 연환산 변동성
        else:
            avg_monthly_variance_dp_sim = 0.0
            avg_annual_volatility_dp_sim = 0.0
        sharpe_ratio_dp_sim = (avg_annual_return_dp_sim - risk_free_rate) / avg_annual_volatility_dp_sim if avg_annual_volatility_dp_sim > 0 else 0 # 샤프 비율

        print("\n--- 최종 DP 시뮬레이션 결과 ---")
        print(f"최종 예상 자본: {final_capital_dp_sim:.2f}")
        print(f"총 수익률: {total_return_dp_sim:.4f}")
        print(f"연 환산 포트폴리오 수익률: {avg_annual_return_dp_sim:.4f}")
        print(f"연 환산 포트폴리오 변동성: {avg_annual_volatility_dp_sim:.4f}")
        print(f"최종 샤프 비율: {sharpe_ratio_dp_sim:.4f}")

        # matplotlib.pyplot과 seaborn을 사용한 시각화 (알고리즘 자체는 아니지만, 결과 표현에 사용)
        plt.figure(figsize=(12, 6))
        plt.plot(range(investment_horizon + 1), capital_path_dp_sim, marker='o')
        plt.title('DP Portfolio Asset Fluctuation (Simulation)', fontsize=15)
        plt.xlabel('Period (Months)', fontsize=12)
        plt.ylabel('Expected Capital', fontsize=12)
        plt.grid(True)
        plt.xticks(range(investment_horizon + 1))
        plt.tight_layout()
        plt.show()

        if not top_3_etfs.empty:
            plt.figure(figsize=(12, 6))
            sns.barplot(x=top_3_etfs.index, y=top_3_etfs['Annual_Return'], palette='viridis')
            plt.title('Annual Returns by Recommended ETF', fontsize=15)
            plt.xlabel('ETF Symbol', fontsize=12)
            plt.ylabel('Annual Return', fontsize=12)
            plt.grid(axis='y')
            plt.tight_layout()
            plt.show()

    else:
        print("\nyfinance 데이터 준비에 실패하여 동적 프로그래밍 단계로 진입할 수 없습니다. 위 오류 메시지를 확인해주세요.")

else:
    print("\n필터링된 상위 ETF가 없으므로 동적 프로그래밍을 진행할 수 없습니다. 조건을 완화하여 다시 시도해 주세요.")
# 4단계: 동적 프로그래밍 정책 기반 백테스팅
print("\n" + "="*50)
print("=== 4단계: 동적 프로그래밍 정책 기반 백테스팅 ===")
print("="*50 + "\n")
if 'policy' in locals() and 'possible_weights' in locals() and not top_3_etfs.empty:
    etf_symbols_for_backtest = top_3_etfs.index.tolist()

    print(f"\n백테스팅을 위해 선택된 ETF의 과거 일별 데이터를 불러오는 중 ({', '.join(etf_symbols_for_backtest)})...")
    end_date_backtest = datetime.now()
    # investment_horizon은 앞에서 입력받은 개월 수 (예: 12개월)에 해당
    start_date_backtest = end_date_backtest - relativedelta(months=investment_horizon)

    print(f"데이터 로드 기간: {start_date_backtest.strftime('%Y-%m-%d')} ~ {end_date_backtest.strftime('%Y-%m-%d')}")

    try:
        # 1단계: ETF 데이터 수집 (yfinance를 사용하여 과거 일별 가격 데이터 다운로드)
        historical_prices_daily = yf.download(etf_symbols_for_backtest, start=start_date_backtest, end=end_date_backtest)['Close']
        if isinstance(historical_prices_daily, pd.Series):
            historical_prices_daily = historical_prices_daily.to_frame()

        historical_prices_daily = historical_prices_daily.dropna() # NaN 값 제거

        if historical_prices_daily.empty or historical_prices_daily.shape[1] != len(etf_symbols_for_backtest):
            print("오류: 백테스팅을 위한 충분한 과거 가격 데이터를 가져오지 못했습니다. 심볼을 확인하거나 데이터 가용성을 확인해주세요.")
            historical_prices_daily = pd.DataFrame()

    except Exception as e:
        print(f"yfinance 데이터 로드 중 오류 발생: {e}")
        historical_prices_daily = pd.DataFrame()

    if not historical_prices_daily.empty:
        initial_capital_backtest = initial_capital
        transaction_cost_rate_backtest = transaction_cost_rate
        risk_free_rate_backtest = risk_free_rate # 무위험 수익률은 백테스팅 성과 측정(샤프 비율)에 사용

        print("\n--- 동적 프로그래밍 정책 기반 포트폴리오 백테스팅 시작 ---")

        current_capital_dp_backtest = initial_capital_backtest # 백테스팅 시뮬레이션 중 현금 보유액
        current_holdings = np.full(num_assets, 0.0) # 각 ETF의 보유 주식 수

        # 4. 탐색 및 정렬 - 초기 가중치에 가장 가까운 상태 찾기
        # 동적 프로그래밍의 `policy` 테이블을 사용하기 위해 초기 상태(가중치)에 해당하는 인덱스를 찾습니다.
        initial_weights_array_for_dp = np.full(num_assets, 1.0 / num_assets)
        initial_weights_tuple_rounded_for_dp = tuple(np.round(initial_weights_array_for_dp, 1))

        try:
            current_weights_idx_dp_backtest = possible_weights.index(initial_weights_tuple_rounded_for_dp)
        except ValueError:
            # 정확히 일치하는 가중치 튜플이 없을 경우, 가장 가까운 것을 찾습니다.
            min_diff = float('inf')
            closest_idx = -1
            for idx, pw in enumerate(possible_weights):
                diff = np.sum(np.abs(np.array(pw) - initial_weights_array_for_dp))
                if diff < min_diff:
                    min_diff = diff
                    closest_idx = idx
            current_weights_idx_dp_backtest = closest_idx

        current_portfolio_weights_dp = np.array(possible_weights[current_weights_idx_dp_backtest]) # 초기 포트폴리오 가중치

        dp_backtest_capital_path = [initial_capital_backtest] # 시간 경과에 따른 총 자산 가치 기록
        dp_backtest_dates = [historical_prices_daily.index[0]] # 날짜 기록

        last_rebalance_month = None # 월별 리밸런싱을 위한 추적 변수

        print(f"DP 백테스팅 기간: {historical_prices_daily.index.min().strftime('%Y-%m-%d')} ~ {historical_prices_daily.index.max().strftime('%Y-%m-%d')}")

        # 백테스팅 루프: 일별 가격 데이터를 순회합니다.
        for i in range(len(historical_prices_daily)):
            current_date = historical_prices_daily.index[i]
            current_prices = historical_prices_daily.iloc[i].values # 현재 날짜의 ETF 가격

            # 월초에 동적 프로그래밍 정책에 따라 리밸런싱 수행
            if current_date.month != last_rebalance_month:
                # 현재 월이 투자 기간(investment_horizon) 내에 있는지 확인
                month_index = (current_date.year - start_date_backtest.year) * 12 + (current_date.month - start_date_backtest.month)

                if month_index < investment_horizon: # 동적 프로그래밍 정책이 정의된 기간 내에서만 적용
                    # 1. 동적 프로그래밍 - 정책 실행 (policy 테이블에서 다음 최적 가중치 조회)
                    optimal_next_w_idx = policy[month_index][current_weights_idx_dp_backtest]

                    if optimal_next_w_idx == -1:
                        next_optimal_weights = current_portfolio_weights_dp # 정책이 없을 경우 이전 가중치 유지
                        print(f"경고: {current_date.strftime('%Y-%m-%d')}에 정책을 찾을 수 없습니다. 직전 가중치 유지.")
                    else:
                        next_optimal_weights = np.array(possible_weights[optimal_next_w_idx]) # 최적 가중치

                    # 현재 포트폴리오의 총 가치 계산 (현금 + 주식 가치)
                    current_portfolio_value = current_capital_dp_backtest + np.sum(current_holdings * current_prices)

                    # 3. 통계적 성과 지표 계산 (거래 비용 계산)
                    # 리밸런싱에 따른 거래 비용 계산
                    cost_amount = calculate_transaction_cost(current_portfolio_weights_dp, next_optimal_weights, current_portfolio_value, transaction_cost_rate_backtest)

                    available_capital_for_rebalance = current_portfolio_value - cost_amount # 거래 비용 제외한 리밸런싱 가능 자본

                    # 목표 보유 금액에 따라 매수할 주식 수 계산
                    target_holdings_value = next_optimal_weights * available_capital_for_rebalance
                    # 0으로 나누는 오류 방지를 위한 처리 (where=current_prices!=0)
                    shares_to_buy = np.divide(target_holdings_value, current_prices, out=np.zeros_like(target_holdings_value), where=current_prices!=0)

                    current_holdings = shares_to_buy # 보유 주식 업데이트
                    current_capital_dp_backtest = 0.0 # 모든 자산이 주식으로 전환되므로 현금 0

                    current_portfolio_weights_dp = next_optimal_weights # 현재 포트폴리오 가중치 업데이트
                    current_weights_idx_dp_backtest = optimal_next_w_idx if optimal_next_w_idx != -1 else current_weights_idx_dp_backtest

                    print(f" {current_date.strftime('%Y-%m-%d')} (월초 리밸런싱): "
                                    f"현재 자산: {current_portfolio_value:.2f}, 적용 가중치: {next_optimal_weights}, "
                                    f"거래 비용: {cost_amount:.2f}")

                last_rebalance_month = current_date.month # 리밸런싱이 일어난 월 기록

            # 일별 포트폴리오 가치 계산 (리밸런싱이 없는 날은 주식 가격 변동만 반영)
            daily_portfolio_value = current_capital_dp_backtest + np.sum(current_holdings * current_prices)
            dp_backtest_capital_path.append(daily_portfolio_value)
            dp_backtest_dates.append(current_date)

        final_capital_dp_backtest = dp_backtest_capital_path[-1]
        total_return_dp_backtest = (final_capital_dp_backtest - initial_capital_backtest) / initial_capital_backtest

        dp_backtest_series = pd.Series(dp_backtest_capital_path, index=dp_backtest_dates)
        dp_backtest_daily_returns = dp_backtest_series.pct_change().dropna()

        # 3. 통계적 성과 지표 계산 - 백테스팅 결과 지표 계산 (수익률, 변동성, MDD, 샤프 비율)
        if not dp_backtest_daily_returns.empty:
            num_trading_days_dp_backtest = len(dp_backtest_daily_returns)
            if num_trading_days_dp_backtest > 0:
                avg_annual_return_dp_backtest = (1 + dp_backtest_daily_returns).prod()**(252/num_trading_days_dp_backtest) - 1 # 연환산 수익률 (252는 연간 평균 거래일 수)
                avg_annual_volatility_dp_backtest = dp_backtest_daily_returns.std() * np.sqrt(252) # 연환산 변동성
                sharpe_ratio_dp_backtest = (avg_annual_return_dp_backtest - risk_free_rate_backtest) / avg_annual_volatility_dp_backtest if avg_annual_volatility_dp_backtest > 0 else 0

                rolling_max_dp = dp_backtest_series.cummax() # 누적 최대 자산 가치
                daily_drawdown_dp = (dp_backtest_series - rolling_max_dp) / rolling_max_dp # 일별 낙폭
                mdd_dp = daily_drawdown_dp.min() * -1 if not daily_drawdown_dp.empty else 0 # 최대 낙폭 (MDD)
            else:
                avg_annual_return_dp_backtest = np.nan
                avg_annual_volatility_dp_backtest = np.nan
                sharpe_ratio_dp_backtest = np.nan
                mdd_dp = np.nan
        else:
            avg_annual_return_dp_backtest = np.nan
            avg_annual_volatility_dp_backtest = np.nan
            sharpe_ratio_dp_backtest = np.nan
            mdd_dp = np.nan


        print("\n--- 동적 프로그래밍 정책 기반 백테스팅 결과 ---")
        print(f"   초기 자본: {initial_capital_backtest:.2f}")
        print(f"   최종 자본: {final_capital_dp_backtest:.2f}")
        print(f"   총 수익률: {total_return_dp_backtest:.4f}")
        print(f"   연환산 수익률: {avg_annual_return_dp_backtest:.4f}")
        print(f"   연환산 변동성: {avg_annual_volatility_dp_backtest:.4f}")
        print(f"   최대 낙폭 (MDD): {mdd_dp:.4f}")
        print(f"   샤프 비율: {sharpe_ratio_dp_backtest:.4f}")

        # matplotlib.pyplot을 사용한 시각화
        plt.figure(figsize=(14, 7))
        plt.plot(dp_backtest_dates, dp_backtest_capital_path, label='DP Portfolio Capital (Backtest)', color='darkblue')
        plt.axhline(y=initial_capital_backtest, color='gray', linestyle='--', label='Initial Capital')
        plt.title('DP Policy-Based Portfolio Backtest - Capital Growth', fontsize=15)
        plt.xlabel('Date', fontsize=12)
        plt.ylabel('Capital', fontsize=12)
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

        print("\n--- 모든 백테스팅 완료 ---")
        print("\n=== 전체 백테스팅 요약 결과 ===")

        summary_data_all = [{
            'Strategy': 'DP Policy Based',
            'Total Return': total_return_dp_backtest,
            'Annualized Return': avg_annual_return_dp_backtest,
            'Annualized Volatility': avg_annual_volatility_dp_backtest,
            'Max Drawdown (MDD)': mdd_dp,
            'Sharpe Ratio': sharpe_ratio_dp_backtest
        }]

        summary_df_all = pd.DataFrame(summary_data_all).set_index('Strategy')
        print(summary_df_all.to_string(float_format=lambda x: f'{x:.4f}'))

        print("\n" + "="*50)

    else:
        print("백테스팅을 위한 유효한 ETF 데이터가 없어 백테스팅을 진행할 수 없습니다.")
else:
    print("추천된 ETF가 없거나 동적 프로그래밍 단계가 실행되지 않아 백테스팅을 진행할 수 없습니다.")
# 5단계: 포트폴리오 시뮬레이션 (미래 예측 시뮬레이션 - 몬테카를로)
print("\n" + "="*50)
print("=== 5단계: 포트폴리오 미래 예측 시뮬레이션 (몬테카를로) ===")
print("="*50 + "\n")

# historical_data_available은 이전 단계에서 yfinance 데이터 로드 성공 여부를 나타내는 플래그
if historical_data_available:
    print("\n--- 몬테카를로 시뮬레이션 파라미터 입력 ---")
    try:
        num_simulations = int(input("시뮬레이션 횟수를 입력해 주세요 (예: 1000): ").strip())
        future_horizon_months = int(input("미래 시뮬레이션 기간을 입력해 주세요 (단위: 개월, 예: 36): ").strip())
        if num_simulations <= 0 or future_horizon_months <= 0:
            raise ValueError("시뮬레이션 횟수와 기간은 양수여야 합니다.")
    except ValueError as e:
        print(f"유효하지 않은 입력입니다: {e}. 기본값 (시뮬레이션 1000회, 기간 36개월)을 사용합니다.")
        num_simulations = 1000
        future_horizon_months = 36

    print(f"\n몬테카를로 시뮬레이션 시작 (횟수: {num_simulations}, 기간: {future_horizon_months}개월)...")

    # 월별 평균 수익률과 월별 공분산 행렬 계산
    monthly_mean_returns = annualized_mean_returns / 12
    monthly_covariance_matrix = covariance_matrix / 12

    # 2. 몬테카를로 시뮬레이션 - 시뮬레이션 경로 초기화
    simulated_paths = np.zeros((num_simulations, future_horizon_months + 1))
    simulated_paths[:, 0] = initial_capital # 모든 시뮬레이션 경로의 시작 자본을 초기 투자 금액으로 설정

    # 2. 몬테카를로 시뮬레이션 - Cholesky 분해를 통한 상관관계 있는 랜덤 변수 생성 준비
    # 포트폴리오의 공분산 행렬을 사용하여 자산 간의 상관관계를 반영한 무작위 수익률을 생성합니다.
    try:
        L = np.linalg.cholesky(monthly_covariance_matrix)
    except np.linalg.LinAlgError:
        print("경고: 공분산 행렬이 양의 반정부호가 아닙니다. 최소한의 노이즈를 추가합니다.")
        # 행렬이 Cholesky 분해 불가능할 경우, 아주 작은 값을 더해 양의 정부호로 만듭니다.
        monthly_covariance_matrix += np.eye(monthly_covariance_matrix.shape[0]) * 1e-6
        L = np.linalg.cholesky(monthly_covariance_matrix)

    # 시뮬레이션에 사용할 포트폴리오 가중치 결정
    # 이전 3단계에서 계산된 최적 포트폴리오 가중치를 사용하거나, 없다면 균등 가중치를 사용합니다.
    if 'current_portfolio_weights' in locals():
        simulation_weights = current_portfolio_weights
    else:
        simulation_weights = np.full(num_assets, 1.0 / num_assets)

    # 3. 통계적 성과 지표 계산 - 시뮬레이션에 사용될 포트폴리오의 월별 예상 수익률과 변동성 계산
    sim_port_mean_return_monthly, sim_port_variance_monthly = calculate_portfolio_metrics(
        simulation_weights, monthly_mean_returns, monthly_covariance_matrix
    )
    sim_port_std_dev_monthly = np.sqrt(sim_port_variance_monthly)

    print(f"\n몬테카를로 시뮬레이션에 사용될 포트폴리오 가중치: {simulation_weights.round(2)}")
    print(f"몬테카를로 시뮬레이션에 사용될 월별 예상 수익률: {sim_port_mean_return_monthly:.4f}")
    print(f"몬테카를로 시뮬레이션에 사용될 월별 예상 변동성: {sim_port_std_dev_monthly:.4f}")

    # 2. 몬테카를로 시뮬레이션 - 각 시뮬레이션 경로 생성
    for sim in range(num_simulations):
        # 정규 분포에서 무작위 월별 수익률 생성 (평균과 표준편차 사용)
        # 이 부분은 단순화를 위해 포트폴리오 전체의 수익률을 직접 정규 분포에서 샘플링합니다.
        # 더 정교하게는, 각 자산의 상관관계를 고려하여 다변량 정규 분포에서 자산별 수익률을 샘플링하고,
        # 이를 가중치에 따라 합산하여 포트폴리오 수익률을 계산할 수 있습니다.
        monthly_returns_path = norm.rvs(loc=sim_port_mean_return_monthly, scale=sim_port_std_dev_monthly, size=future_horizon_months)

        for month in range(future_horizon_months):
            # 이전 월의 자본에 월별 수익률을 곱하여 현재 월의 자본 계산
            simulated_paths[sim, month+1] = simulated_paths[sim, month] * (1 + monthly_returns_path[month])

    final_capitals = simulated_paths[:, -1] # 각 시뮬레이션의 최종 자본

    # 3. 통계적 성과 지표 계산 - 최종 자본의 통계적 요약 (평균, 최대, 최소, 분위수)
    percentiles = [5, 25, 50, 75, 95]
    percentile_values = np.percentile(final_capitals, percentiles)

    print("\n--- 몬테카를로 시뮬레이션 최종 결과 요약 ---")
    print(f"총 시뮬레이션 횟수: {num_simulations}")
    print(f"시뮬레이션 기간: {future_horizon_months} 개월")
    print(f"초기 자본: {initial_capital:.2f}")
    print(f"평균 최종 자본: {np.mean(final_capitals):.2f}")
    print(f"최대 최종 자본: {np.max(final_capitals):.2f}")
    print(f"최소 최종 자본: {np.min(final_capitals):.2f}")
    print("\n최종 자본 분위수:")
    for p, val in zip(percentiles, percentile_values):
        print(f"   {p}th 백분위수: {val:.2f}")

    # matplotlib.pyplot을 사용한 시각화
    plt.figure(figsize=(14, 8))
    # 전체 시뮬레이션 경로 중 일부(최대 100개)를 회색으로 플로팅
    for i in range(min(num_simulations, 100)):
        plt.plot(simulated_paths[i], color='lightgray', alpha=0.5, linewidth=0.5)

    # 50th, 25th, 75th 백분위수 경로를 별도로 플로팅하여 경향성 시각화
    median_path = np.percentile(simulated_paths, 50, axis=0)
    plt.plot(median_path, color='blue', linewidth=2, label='50th Percentile')

    q1_path = np.percentile(simulated_paths, 25, axis=0)
    q3_path = np.percentile(simulated_paths, 75, axis=0)
    plt.plot(q1_path, color='green', linestyle='--', label='25th Percentile')
    plt.plot(q3_path, color='red', linestyle='--', label='75th Percentile')


    plt.title(f'Portfolio Future Asset Prediction (Monte Carlo Simulation - {num_simulations} runs)', fontsize=16)
    plt.xlabel('Period (Months)', fontsize=12)
    plt.ylabel('Expected Capital', fontsize=12)
    plt.axhline(y=initial_capital, color='gray', linestyle='--', label='Initial Capital')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

else:
    print("과거 데이터가 없어 몬테카를로 시뮬레이션을 진행할 수 없습니다.")


# 6단계: 포트폴리오 관리 및 모니터링 (사후 관리)
print("\n" + "="*50)
print("=== 6단계: 포트폴리오 관리 및 모니터링 (사후 관리) ===")
print("="*50 + "\n")

# 필요한 변수들이 이전 단계에서 정의되었는지 확인
if historical_data_available and 'policy' in locals() and 'possible_weights' in locals():
    print("\n--- 현재 포트폴리오 상태 입력 ---")
    current_portfolio = {} # 사용자가 보유한 ETF별 수량을 저장
    for symbol in etf_symbols:
        while True:
            try:
                shares = int(input(f"현재 보유하고 있는 {symbol}의 수량을 입력해 주세요 (정수, 없으면 0): ").strip())
                if shares >= 0:
                    current_portfolio[symbol] = shares
                    break
                else:
                    print("수량은 0 또는 양의 정수여야 합니다.")
            except ValueError:
                print("유효하지 않은 입력입니다. 정수로 입력해주세요.")

    try:
        # 1단계: ETF 데이터 수집 (yfinance를 사용하여 최신 가격 가져오기)
        latest_prices = yf.download(etf_symbols, period='1d')['Close'].iloc[-1]
        print(f"\n현재 ETF 가격: {latest_prices.to_dict()}")
    except Exception as e:
        print(f"오류: 최신 ETF 가격을 가져오는 데 실패했습니다: {e}")
        print("포트폴리오 관리를 위해 가격을 수동으로 입력해야 합니다.")
        latest_prices = pd.Series(index=etf_symbols, dtype=float)
        for symbol in etf_symbols:
            while True:
                try:
                    price = float(input(f"현재 {symbol}의 최신 가격을 입력해 주세요: ").strip())
                    if price > 0:
                        latest_prices[symbol] = price
                        break
                    else:
                        print("가격은 양수여야 합니다.")
                except ValueError:
                    print("유효하지 않은 입력입니다. 숫자로 입력해주세요.")

    print("\n--- 현재 포트폴리오 분석 ---")
    current_total_value = 0
    current_individual_values = {} # 각 ETF의 현재 시장 가치
    for symbol, shares in current_portfolio.items():
        if symbol in latest_prices and not pd.isna(latest_prices[symbol]):
            value = shares * latest_prices[symbol]
            current_individual_values[symbol] = value
            current_total_value += value
        else:
            current_individual_values[symbol] = 0
            print(f"경고: {symbol}의 가격 데이터를 가져올 수 없어 해당 ETF의 가치 계산에서 제외됩니다.")

    if current_total_value == 0:
        print("현재 포트폴리오의 총 가치가 0입니다. 리밸런싱 제안을 할 수 없습니다.")
        current_weights = np.zeros(num_assets)
    else:
        current_weights = np.array([current_individual_values[s] / current_total_value for s in etf_symbols]) # 현재 포트폴리오 가중치

    print(f"총 포트폴리오 가치: {current_total_value:.2f} KRW")
    print("현재 ETF별 가치:")
    for symbol, value in current_individual_values.items():
        print(f"   {symbol}: {value:.2f} KRW")
    print(f"현재 ETF별 가중치: {current_weights.round(3)}")

    # 3. 통계적 성과 지표 계산 - 현재 포트폴리오의 예상 성과 계산
    if np.sum(current_weights) > 0:
        current_port_return, current_port_variance = calculate_portfolio_metrics(
            current_weights, annualized_mean_returns, covariance_matrix
        )
        current_port_volatility = np.sqrt(current_port_variance)
        current_port_sharpe = (current_port_return - risk_free_rate) / current_port_volatility if current_port_volatility > 0 else np.nan

        print("\n--- 현재 포트폴리오 예상 성과 (연간 기준) ---")
        print(f"예상 수익률: {current_port_return:.4f}")
        print(f"예상 변동성: {current_port_volatility:.4f}")
        print(f"예상 샤프 비율: {current_port_sharpe:.4f}")
    else:
        print("\n현재 포트폴리오에 보유 자산이 없어 예상 성과를 계산할 수 없습니다.")

    print("\n--- 리밸런싱 권장 사항 ---")
    # 4. 탐색 및 정렬 - 현재 가중치와 가장 유사한 DP 정책 상태 인덱스 찾기
    current_w_array_rounded = np.round(current_weights, 1) # 반올림하여 이산 가중치 상태와 매칭 시도

    min_diff_for_idx = float('inf')
    closest_idx_for_policy = -1
    for idx, pw in enumerate(possible_weights):
        diff = np.sum(np.abs(np.array(pw) - current_w_array_rounded))
        if diff < min_diff_for_idx:
            min_diff_for_idx = diff
            closest_idx_for_policy = idx

    if closest_idx_for_policy != -1:
        # 1. 동적 프로그래밍 - 정책 기반 리밸런싱 제안
        # DP policy 테이블에서 현재 상태(closest_idx_for_policy)에 대한 다음(0개월 시점, 즉시 리밸런싱) 최적 가중치 조회
        target_weights_idx = policy[0][closest_idx_for_policy] # policy[0]는 현재 시점 (0개월)의 최적 정책
        if target_weights_idx != -1:
            target_weights = np.array(possible_weights[target_weights_idx])
            print(f"현재 포트폴리오 가중치: {current_weights.round(3)}")
            print(f"DP 모델의 권장 목표 가중치 (첫 리밸런싱): {target_weights.round(3)}")

            if current_total_value > 0:
                actions = []
                # 리밸런싱을 위한 매수/매도 제안 계산
                for j, symbol in enumerate(etf_symbols):
                    current_value_asset = current_individual_values[symbol]
                    target_value_asset = target_weights[j] * current_total_value # 목표 자산 가치

                    price_asset = latest_prices[symbol]
                    if pd.isna(price_asset) or price_asset == 0:
                        print(f"경고: {symbol}의 현재 가격을 알 수 없어 해당 ETF의 매수/매도 제안을 건너뜝니다.")
                        continue

                    current_shares = current_portfolio[symbol]
                    target_shares = target_value_asset / price_asset # 목표 주식 수

                    share_diff = target_shares - current_shares # 목표 주식 수와 현재 주식 수의 차이

                    # 거래 비용을 고려하여 의미 있는 리밸런싱만 제안
                    if abs(share_diff * price_asset) > (current_total_value * transaction_cost_rate):
                        if share_diff > 0:
                            # 매수 제안
                            cost_of_buy = abs(share_diff * price_asset) * transaction_cost_rate
                            actions.append(f"   {symbol}: 약 {share_diff:.2f}주 매수 (금액: {abs(share_diff * price_asset):.2f} KRW, 예상 비용: {cost_of_buy:.2f} KRW)")
                        elif share_diff < 0:
                            # 매도 제안
                            revenue_of_sell = abs(share_diff * price_asset)
                            cost_of_sell = revenue_of_sell * transaction_cost_rate
                            actions.append(f"   {symbol}: 약 {abs(share_diff):.2f}주 매도 (금액: {revenue_of_sell:.2f} KRW, 예상 비용: {cost_of_sell:.2f} KRW)")

                if actions:
                    print("\n다음과 같이 리밸런싱을 고려해 볼 수 있습니다 (거래 비용 포함):")
                    for action in actions:
                        print(action)

                    # 3. 통계적 성과 지표 계산 (거래 비용 계산)
                    total_estimated_rebalancing_cost = calculate_transaction_cost(current_weights, target_weights, current_total_value, transaction_cost_rate)
                    print(f"\n총 예상 리밸런싱 거래 비용: {total_estimated_rebalancing_cost:.2f} KRW")
                    print("리밸런싱 후 포트폴리오 가치는 거래 비용만큼 감소합니다.")
                else:
                    print("현재 포트폴리오는 목표 가중치와 큰 차이가 없어 리밸런싱이 필요하지 않습니다.")
            else:
                print("현재 포트폴리오에 자산이 없습니다. 목표 가중치에 따라 자산을 매수하여 포트폴리오를 구성할 수 있습니다.")
                # 초기 포트폴리오 구성 제안
                for j, symbol in enumerate(etf_symbols):
                    price_asset = latest_prices[symbol]
                    if pd.isna(price_asset) or price_asset == 0:
                        print(f"경고: {symbol}의 현재 가격을 알 수 없어 해당 ETF의 매수 제안을 건너뜝니다.")
                        continue

                    target_value_asset_for_initial = target_weights[j] * initial_capital # 초기 자본 기준으로 목표 가치 계산
                    shares_to_buy_initial = target_value_asset_for_initial / price_asset
                    cost_of_buy_initial = target_value_asset_for_initial * transaction_cost_rate
                    print(f"   {symbol}: 약 {shares_to_buy_initial:.2f}주 매수 (금액: {target_value_asset_for_initial:.2f} KRW, 예상 비용: {cost_of_buy_initial:.2f} KRW)")

                # 3. 통계적 성과 지표 계산 (초기 포트폴리오 구성 비용 계산)
                total_initial_cost = calculate_transaction_cost(np.zeros(num_assets), target_weights, initial_capital, transaction_cost_rate)
                print(f"\n총 예상 초기 포트폴리오 구성 비용: {total_initial_cost:.2f} KRW")

        else:
            print("DP 정책에서 현재 가중치에 대한 다음 기간의 최적 가중치를 찾을 수 없습니다.")
    else:
        print("현재 포트폴리오 가중치에 가장 가까운 DP 정책 상태를 찾을 수 없습니다.")
else:
    print("과거 데이터가 없거나 동적 프로그래밍 단계가 실행되지 않아 포트폴리오 관리 및 모니터링을 진행할 수 없습니다.")

print("\n" + "="*50)
print("=== 프로그램 종료 ===")
print("="*50 + "\n")
```
